````python
import numpy as np
import matplotlib.pyplot as plt
import control.matlab as mt
from warnings import filterwarnings

filterwarnings('ignore')
````


# Bilinear low pass filter transfer function

 $\frac {1}{s+1}$

````python
num=[1]
den=[1,1]

sys=mt.tf(num,den)
sys
````

````python
mag,phase,omega=mt.bode(sys)
````

Bilinear high pass filter transfer function

$\frac{s}{s+1}$

````python
num=[1,0]
den=[1,1]
sys=mt.tf(num,den)
mag,phase,omega=mt.bode(sys)
````

# Second order(Biquad) Low pass filter tranfer function

$\frac{4}{s^2+8s+1}$

````python
num=[4]
den=[1,8,1]

sys=mt.tf(num,den)
mag,phase,omega=mt.bode(sys)
````
